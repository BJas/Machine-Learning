```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

img1 = cv2.imread('/content/photo.jpg',cv2.IMREAD_GRAYSCALE)
img_resized1 = cv2.resize(img1,(400,500))
cv2_imshow(img_resized1)
```

```
#Rotate image
(h, w) = img_resized1.shape[:2]
# calculate the center of the image
center = (w / 2, h / 2)
 
angle90 = 90
angle180 = 180
angle270 = 270
 
scale = 1.0

M = cv.getRotationMatrix2D(center, angle270, scale)
rotated270 = cv.warpAffine(img_resized1, M, (w, h))

plt.imshow(rotated270, 'gray')
```
